```python
import pandas as pd

cand = pd.read_csv("benchmark/candidate_pool_for_labeling.csv")

# Keep only labeled rows
qrels = cand[cand["relevance"].notna()].copy()

# If relevance was entered as text, force numeric
qrels["relevance"] = pd.to_numeric(qrels["relevance"], errors="coerce")
qrels = qrels[qrels["relevance"].isin([0, 1, 2])]

qrels = qrels[["query_id", "doc_id", "relevance", "notes"]].copy()
qrels.to_csv("benchmark/qrels.csv", index=False)

print("qrels rows:", len(qrels))
print(qrels.head())
```
